```python
two_bands = dc.load(product='lc08', resolution=(-0.0025, 0.00250), measurements=['red', 'nir'])
px = dc.load(product = 'lc08', resolution=(-0.0025, 0.00250), measurements=['quality'])
```

```python
red = two_bands.red.where(two_bands.red != two_bands.red.attrs['nodata'])
nir = two_bands.nir.where(two_bands.nir != two_bands.nir.attrs['nodata'])

cloud_free = masking.make_mask(px, water_confidence=True).quality
ndvi = ((nir - red) / (nir + red)).where(cloud_free)

ndvi.shape
```

```python
ndvi.plot(col='time', col_wrap=5)
```

```python
red = two_bands.red.where(two_bands.red != two_bands.red.attrs['nodata'])
nir = two_bands.nir.where(two_bands.nir != two_bands.nir.attrs['nodata'])

cloud_free = masking.make_mask(px, cirrus_confidence='Low').quality
ndvi = ((nir - red) / (nir + red)).where(cloud_free)

ndvi.shape
```

```python
ndvi.plot(col='time', col_wrap=5)
```

```python
mostly_cloud_free = cloud_free.sum(dim=('longitude','latitude')) > (0.75 * cloud_free.size / cloud_free.time.size)
mostly_good_ndvi = ndvi.where(mostly_cloud_free).dropna('time', how='all')
mostly_good_ndvi.plot(col='time', col_wrap=5)
```
