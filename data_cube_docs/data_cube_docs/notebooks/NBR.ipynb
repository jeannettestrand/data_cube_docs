```python
# This Notebook loads two datasets from memory, calculates delta NBR on two results and plots the results.

# The NBR ration highlights areas that have burned to index the severity of a burn using Landsat TM imagery.
# NBR = (nir - swir) / (nir + swir)
```

```python
import datacube
from datacube.storage import masking

dc = datacube.Datacube()
```

```python
# Declare some 'constants' for the program
p = 'LT05_scene'
res = (0.0025, -0.0025)
m = ['swir1', 'nir']
preStart = '1998'
postTime = '1999'
```

```python
dc.list_measurements()
```

```python
dc.find_datasets(product = p)
```

```python
pre = dc.load(product = p
            , measurements = m
            , time = ('2001-08-08', '2001-08-09')
            , resolution = res
            )
post = dc.load(product = p
            , measurements = m
            , time = ('2001-08-24', '2001-08-25')
            , resolution = res
            )
pre
```

```python
# Isolate all valid data in the the two bands of interest
preSwir = pre.swir1.where(pre.swir1 != pre.swir1.attrs['nodata'])
preNir = pre.nir.where(pre.nir != pre.nir.attrs['nodata'])

postSwir = post.swir1.where(post.swir1 != post.swir1.attrs['nodata'])
postNir = post.nir.where(post.nir != post.nir.attrs['nodata'])
```

```python
#Calculate pre and post NBR
preNBR = ((preNir - preSwir) / (preNir + preSwir))
postNBR = ((postNir - postSwir) / (postNir + postSwir))
```
